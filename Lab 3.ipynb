# Lab 3: Gradient Descent and Optimization

#### Implement a basic gradient descent algorithm for a simple function.
#### Visualize the convergence of the algorithm over iterations.

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function to optimize (example: quadratic function)
def function_to_optimize(x):
    return x**2 + 5*x + 10
```

```python
# Define the gradient of the function
def gradient(x):
    return 2*x + 5
```

```python
# Gradient Descent Algorithm
def gradient_descent(learning_rate, iterations):
    x = 0  # Initial guess for the minimum
    x_values = []
    y_values = []
    
    for i in range(iterations):
        gradient_value = gradient(x)
        x -= learning_rate * gradient_value
        
        # Store the values for visualization
        x_values.append(x)
        y_values.append(function_to_optimize(x))
        
    return x_values, y_values
```

```python
# Parameters
learning_rate = 0.1
iterations = 50
```

```python
# Run Gradient Descent
x_values, y_values = gradient_descent(learning_rate, iterations)
```

```python
# Plot the convergence
plt.figure(figsize=(10, 5))
plt.plot(range(iterations), y_values, marker='o', linestyle='-')
plt.title('Convergence of Gradient Descent')
plt.xlabel('Iterations')
plt.ylabel('Function Value')
plt.grid(True)
plt.show()
```
